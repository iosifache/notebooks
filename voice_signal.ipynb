# Voice Signal

## Libraries Import

```python
import sys
import wave

import matplotlib.pyplot as pyplot
import numpy as np
from scipy.fft import fft, fftfreq
from scipy.io import wavfile
```

## Configuration

```python
SOUNDS_FOLDER = "sounds/"
A_SOUND_FILENAME = SOUNDS_FOLDER + "a.wav"
SARE_SOUND_FILENAME = SOUNDS_FOLDER + "sare.wav"
PHRASE_SOUND_FILENAME = SOUNDS_FOLDER + "phrase.wav"
```

## Representation in the Time Domain

```python
def plot_sound_in_time(filename: str) -> None:
    # Read the frame of the signal
    wav_file = wave.open(filename)
    raw_signal = wav_file.readframes(-1)
    frames = np.frombuffer(raw_signal, dtype='int16')

    # Create the time axis
    rate = wav_file.getframerate()
    samples_count = len(frames)
    time = np.linspace(0, samples_count / rate, num = samples_count)

    # Plot the signal
    pyplot.figure()
    pyplot.title("Time Representation of the Signal '{}''".format(filename))
    pyplot.xlabel("Time (s)")
    pyplot.ylabel("Amplitude")
    pyplot.plot(time, frames)
    pyplot.show()
```

```python
plot_sound_in_time(A_SOUND_FILENAME)
plot_sound_in_time(SARE_SOUND_FILENAME)
plot_sound_in_time(PHRASE_SOUND_FILENAME)
```

## Representation in the Frequency Domain

```python
def plot_sound_in_frequency(filename: str) -> None:
    # Read the frame of the signal
    wav_file = wave.open(filename)
    raw_signal = wav_file.readframes(-1)
    frames = np.frombuffer(raw_signal, dtype='int16')
    transform = fft(frames)
    transform = transform[0:len(transform) // 2]

    # Create the time axis
    rate = wav_file.getframerate()
    frequency = np.linspace(0, rate, num = len(transform))

    # Plot the signal
    pyplot.figure()
    pyplot.title("Frequency Representation of the Signal '{}''".format(filename))
    pyplot.xlabel("Frequency (Hz)")
    pyplot.ylabel("Amplitude")
    pyplot.plot(frequency, 2 / len(transform) * np.abs(transform))
    pyplot.show()
```

```python
plot_sound_in_frequency(A_SOUND_FILENAME)
plot_sound_in_frequency(SARE_SOUND_FILENAME)
plot_sound_in_frequency(PHRASE_SOUND_FILENAME)
```
